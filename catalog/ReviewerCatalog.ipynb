# Add a Reviewer Suite to Catalog


1. Set up repo
    Registered repos must satisfy the following:
    - Public
    - Has a description
    - Structured like so (see [AnnoMateTemplate](https://github.com/getzlab/AnnoMateTemplate))
        ```
        MyReviewers/
            MyReviewers/
                Reviewers/
                    Reviewer1.py
                    Reviewer2.py
                    ...
                AppComponents/
                    AppComponent1.py
                    AppComponent2.py
                    ...
        ```
   - `setup.py` file

1. Fork AnnoMate and go to `example_notebooks/ReviewerCatalog.py`
    1. Add the github url of your reviewer repo to the variable `registered_reviewer_repos`
    1. Go to `catalog/ReviewerCatalog.ipynb` and run `gen_reviewer_catalog()` (below)
    1. Check `catalog.tsv`
    
1. Commit and submit a pull request
