```python
workspace = 'broad-getzlab-pcyc-terra/Getz_Wu_CLL_Pharmacyclics_Exomes_Analysis_8_tumor_only'
pcyc_wm = dalmatian.WorkspaceManager(workspace)
pcyc_wm_pairs_df = pcyc_wm.get_pairs().iloc[:5]
display(pcyc_wm_pairs_df.head())
```

# Download rdata to local

```python
import os
import dalmatian

def download_data(file_to_download_path, full_local_path):
    dalmatian.getblob(file_to_download_path).download_to_filename(full_local_path)
```

```python
rdata_dir = f'/Users/cchu/Desktop/Methods/JupyterReviewer/data/Purity_Reviewer_Tutorial/local_rdata'
if not os.path.isdir(rdata_dir):
    os.mkdir(rdata_dir)
    
for pair_id, r in pcyc_wm_pairs_df.iterrows():
    absolute_rdata_gsurl = r['absolute_rdata']
    local_absolute_rdata_fn = f'{rdata_dir}/{pair_id}.absolute.rdata'
    if not os.path.exists(local_absolute_rdata_fn):
        download_data(absolute_rdata_gsurl, local_absolute_rdata_fn)
    pcyc_wm_pairs_df.loc[pair_id, 'local_absolute_rdata'] = local_absolute_rdata_fn
    
```

# Set up purity reviewer

```python
review_data_fn = '/Users/cchu/Desktop/Methods/JupyterReviewer/data/Purity_Reviewer_Tutorial/review_data.7.pkl'

matched_reviewer = MatchedPurityReviewer()
matched_reviewer.set_review_data(data_pkl_fn = review_data_fn, 
                                 description='testing matched purity reviewer', 
                                 df=pcyc_wm_pairs_df, # optional if directory above already exists. 
                                 index=pcyc_wm_pairs_df.index,
                                 preprocess_data_dir= '/Users/cchu/Desktop/Methods/JupyterReviewer/data/Purity_Reviewer_Tutorial_3',
                                 acs_col='alleliccapseg_tsv', 
                                 maf_col='mutation_validator_validated_maf',
                                 rdata_fn_col='local_absolute_rdata',
                                 reload_cnp_figs=False,
                                 reload_mut_figs=False,
                                )
matched_reviewer.set_review_app(sample_info_cols=['absolute_highres_plot', 'alleliccapseg_plot', 'wxs_purity'],
                                acs_col='alleliccapseg_tsv', 
                                maf_col='mutation_validator_validated_maf')

matched_reviewer.set_default_review_data_annotations_configuration()
matched_reviewer.set_default_autofill()
```

```python
matched_reviewer.run(port=8070, review_data_table_df=pcyc_wm_pairs_df[['wxs_purity', 'participant']])
```
