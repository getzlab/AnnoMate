# Add a Reviewer Suite to Registry


1. Set up repo
    Registered repos must satisfy the following:
    - Public
    - Has a description
    - Structured like so (see [JupyterReviewerTemplate](https://github.com/getzlab/JupyterReviewerTemplate))
        ```
        MyReviewers/
            MyReviewers/
                Reviewers/
                    Reviewer1.py
                    Reviewer2.py
                    ...
                AppComponents/
                    AppComponent1.py
                    AppComponent2.py
                    ...
        ```
   - `setup.py` file

1. Fork JupyterReviewer and go to `example_notebooks/ReviewerRegistry.py`
    1. Add the github url of your reviewer repo to the variable `registered_reviewer_repos`
    1. Go to `example_notebooks/ReviewerRegistry.ipynb` and run `gen_reviewer_registry()` (below)
    1. Check `registry.tsv`
    
1. Commit and submit a pull request
