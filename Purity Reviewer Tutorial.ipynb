```python
data_path = '/home/cchu/cgaprojects_ibm_tAML_analysis/data/test_getzlab-JupyterReviewer/pcyc_data'
rdata_dir = f'{data_path}/absolute_rdata'

import os

def download_rdata(gsurl, local_dir):
    out_path = os.path.join(local_dir, os.path.basename(gsurl))
    if not os.path.exists(out_path):
        dalmatian.getblob(gsurl).download_to_filename(out_path)
    return out_path


pcyc_wm_pairs_df['local_absolute_rdata'] = pcyc_wm_pairs_df['absolute_rdata'].apply(lambda gsurl: download_rdata(gsurl, rdata_dir))

```
